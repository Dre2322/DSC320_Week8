```python
# 1. Solving a System of Linear Equations

import numpy as np

# create an array of all constants on the left side of the equation
A = np.array([[27, -10, 4, -29],[-16, 5 , -2, 18],[-17, 4, -2, 20], [-7, 2, -1, 8]])

# create an array of a constants on the right side of the equation
b = np.array([[1], [-1], [0], [1]])

# solve by taking inverse of A and multiplying it by b
x = np.linalg.solve(A, b)

# print results
print(x)

```

```python
# 2. Visualizing a Linear Transformation
# a

import numpy as np

# linear transformation that takes in a vector v in R2 and outputs the product
A = np.array([[np.sqrt(2)/2, -np.sqrt(2)/2],
              [np.sqrt(2)/2, np.sqrt(2)/2]])

# vector v in R2
x = np.array([[2], [-2]])

# compute the product of A and x
T = np.dot(A, x)

# print results
print(T)
```

```python
# b

import numpy as np
import matplotlib.pyplot as plt

# creating a grid of points
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)

xx, yy = np.meshgrid(x, y)

# creating a transformation matrix T
T = np.array([[-1, 0], 
              [0, -1]])

# structure the points of the grid to be able to apply the matrix to them
xy =  np.vstack([xx.flatten(), yy.flatten()])

# apply the transformation matrix to the grid
trans = T @ xy

# reshape the grid
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

# plot the before and after of the transformed grid
f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)

plt.show()
```
